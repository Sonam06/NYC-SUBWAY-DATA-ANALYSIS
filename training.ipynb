```python
rain_entries = data_set.groupby('rain')['ENTRIESn_hourly'].mean()
rain_entries

## IT INDICATES THAT ENTRIES IN SUBWAY IS ALMOST SAME WHILE RAINING OR NOT RAINING.
```

```python
%pylab inline

ax = rain_entries.plot(kind='bar',
                      title='Entries in Subway when raining v/s not raining',
                      grid=True,
                      legend=True,
                      color='black')
ax.set(xlabel="Not Raining V/s Raining", ylabel="Number of people entering the subway")
ax.set_ylim([1050,1150])
```

```python
rain_exit = data_set.groupby('rain')['EXITSn_hourly'].mean()
rain_exit
## IT INDICATES THAT EXITS IN SUBWAY WHILE RAINING IS ALMOST THE SAME AS WHEN IT'S NOT RAINING.
## ALSO, THE AVERAGE OF PEOPLE ENTERING THE SUBWAY WHILE IT WAS NOT RAINING IS LESS THAN WHEN IT WAS RAINING.
## BELOW THAT EXITS IN THE SUBWAY WHILE IT WAS NOT RAINING IS LESS THAN WHEN IT WAS RAINING.
```

```python
ax = rain_exit.plot(kind='bar',
                      title='Exit in Subway when raining v/s not raining',
                      grid=True,
                      legend=True,
                      color='blue')
ax.set(xlabel="Not Raining V/s Raining", ylabel="Number of people exiting the subway")
ax.set_ylim([850,900])
```

```python
def time_of_day(time):
    if time >= 5 and time < 12:
        return 'Morning'
    elif time >= 12 and time < 17:
        return 'Afternoon'
    elif time >= 17 and time < 24:
        return 'Evening'
    else:
        return 'Night'
    
data_set['Day_Of_Time'] = data_set['Hour'].apply(time_of_day)
data_set['Day_Of_Time'].head(10)
```

```python
time_of_day_rain_probability = data_set.groupby('Day_Of_Time')['rain'].mean()*100
time_of_day_rain_probability

#PROBABILITY OF RAINING DURING FOUR PARTS OF THE DAY IS ALMOST THE SAME.
```

```python
entry_day_of_time = data_set.groupby(['rain', 'Day_Of_Time'],sort=False)['ENTRIESn_hourly'].mean()
entry_day_of_time
```

```python
entry_day_of_time.loc[0.0].plot(kind='bar',
                               grid=True,
                               position=0,
                               legend=True,
                               label="Not Raining",
                               width=0.3,
                               color="red")

entry_day_of_time.loc[1.0].plot(kind='bar',
                               grid=True,
                               position=1,
                               legend=True,
                               label="Raining",
                               width=0.3,
                               color="blue")

## SUBWAY USED MOSTLY DURING AFTERNOON AND EVENING.
```

```python
hourly_rain = data_set.groupby('Hour')['rain'].mean()*100
hourly_rain

## IT INDICATES THAT THE PROBABILITY OF RAINING ANY TIME OF THE DAY IS ALMOST EQUAL IN THE ANALYSED LOCATION.
```

```python
entry_hourly = data_set.groupby(['rain', 'Hour'])['ENTRIESn_hourly'].mean()
```

```python
entry_hourly.loc[0.0].plot(kind='bar',
                          grid=True,
                          position=0,
                          label="Not Raining",
                          legend=True, 
                          width=0.4,
                          color="red")

entry_hourly.loc[1.0].plot(kind='bar',
                          grid=True,
                          position=1,
                          label="Raining",
                          legend=True, 
                          width=0.4,
                          color="blue")
```

```python
## RED INDICATES NOT RAINING AND BLUE INDICATES RAINING. 
## GRAPH INDICATES THAT ENTRIES IN SUBWAY IS MAXIMUM AT THE RATE OF 20:00.
## AND AS WE HAVE OBSERVE ENTRIES IS DIRECTLY PROPORTIONAL TO EXITS.
## THEREFORE EXITS WILL ALSO BE MAXIMUM AT THE RATE OF 20:00.
```

```python
entry_count_day_of_time = data_set.groupby(['rain', 'Day_Of_Time'],sort=False)['ENTRIESn_hourly'].count()
entry_count_day_of_time
```

```python
entry_count_day_of_time.loc[0.0].plot(kind='bar',
                               grid=True,
                               position=0,
                               legend=True,
                               label="Not Raining",
                               width=0.2,
                               color="pink")

entry_count_day_of_time.loc[1.0].plot(kind='bar',
                               grid=True,
                               position=1,
                               legend=True,
                               label="Raining",
                               width=0.2,
                               color="blue")
```

```python
## THIS PLOT SHOWS THAT THE DATA WE HAVE FOR NOT RAINING IS ALMOST TWICE AS MUCH THE DATA OF WHEN IT WAS RAINING.
## AND STILL IN EVERY ASPECT SUBWAY IS USED ALMOST EQUALLY WHEN IT WAS RAINING OR NOT RAINING.
```

```python
from datetime import datetime as dt

def date_to_day(date):
    return dt.strptime(date, '%Y-%m-%d').strftime('%A')

data_set['DAYn'] = data_set['DATEn'].apply(date_to_day)
```

```python
exits_per_day = data_set.groupby(['DAYn'],sort=False)['EXITSn_hourly'].mean()
exits_per_day
```

```python
ax = exits_per_day.plot(kind='bar',
                      title='Exits per day in subway',
                      grid=True,
                      legend=True,
                      color='blue')
ax.set(xlabel="Days", ylabel="Number of people exiting the subway")
```

```python
## SHOWS THAT ON HOLIDAYS i.e. ON SATURDAY AND SUNDAY, PEOPLE USE SUBWAY LESS THAN REGULAR DAYS.
```

```python
entries_per_day = data_set.groupby(['DAYn'],sort=False)['ENTRIESn_hourly'].mean()
entries_per_day
```

```python
ax = entries_per_day.plot(kind='bar',
                      title='Entries per day in subway',
                      grid=True,
                      legend=True,
                      color='red')
ax.set(xlabel="Days", ylabel="Number of people entering the subway")
```
